```python
img=img.resize((28,28))
img=np.array(img)
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
img=255-img
for x in img:
  num_str=""
  for i in x:
    num_str=num_str+str(i)+"\t"
  print(num_str)
```

```python
img=img/255.0
for x in img:
  num_str=""
  for i in x:
    num_str=num_str+str(i)+"\t"
  print(num_str)
```

```python
img=img.reshape((1,28,28,1))
img.shape
```

```python
from keras.models import load_model
```

```python
model=load_model('/content/drive/MyDrive/창의적종합설계 2분반/Predict_Model.h5')
```

```python
predict=model.predict_on_batch(img)
mypred=np.argmax(predict, axis=1)
```

```python
plt.figure(figsize = (12,5))

plt.subplot(1,2,1)
plt.imshow(img.reshape(28, 28), 'gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.stem(predict[0])
plt.show()

print('Prediction : {}'.format(mypred[0]))
```
