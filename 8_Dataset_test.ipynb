```python
from keras.models import load_model
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
img =Image.open("/content/drive/MyDrive/창의적종합설계 2분반/test_5.jpg").convert("L")
plt.imshow(img)
plt.show()
```

```python
img=img.resize((100,100))
img=np.array(img)
print(img.shape)
```

```python
img=255-img
img=img/255.0
img=img.reshape((1,100,100,1))
print(img.shape)
```

```python
model = load_model('/content/drive/MyDrive/창의적종합설계 2분반/ox_Model.h5')
```

```python
predict=model.predict_on_batch(img)
mypred=np.argmax(predict, axis=1)
print(mypred)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.imshow(img.reshape(100,100),'gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.stem(predict[0])
plt.show()
```
